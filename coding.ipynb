```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
df_std1 = sc.fit_transform(df2[['Kematian_anak','Jumlah_fertiliti']].astype(float))
df_std2 = sc.fit_transform(df2[['Kematian_anak','Harapan_hidup']].astype(float))
df_std3 = sc.fit_transform(df2[['Kematian_anak','Pendapatan']].astype(float))
df_std4 = sc.fit_transform(df2[['Ekspor','Impor']].astype(float))
df_std5 = sc.fit_transform(df2[['Ekspor','Pendapatan']].astype(float))
df_std6 = sc.fit_transform(df2[['Pendapatan','Jumlah_fertiliti']].astype(float))
df_std7 = sc.fit_transform(df2[['Pendapatan','Harapan_hidup']].astype(float))
df_std8 = sc.fit_transform(df2[['Pendapatan','GDPperkapita']].astype(float))
df_std9 = sc.fit_transform(df2[['Jumlah_fertiliti','Harapan_hidup']].astype(float))
df_std10 = sc.fit_transform(df2[['GDPperkapita','Harapan_hidup']].astype(float))
df_std = sc.fit_transform(df2.astype(float))
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
wcss = []
for i in range(1,11):
    kmeans1 = KMeans(n_clusters=i, init='k-means++', random_state=3)
    kmeans1.fit(df_std)
    wcss.append(kmeans1.inertia_)

plt.plot(range(1,11), wcss)
plt.title('Elbow method')
plt.xlabel('n_clusters')
plt.ylabel('wcss')

plt.show()
```

```python
aa=4
b=['blue','red','green','pink']
c=['Kematian_anak','Jumlah_fertiliti']
kmeans1 = KMeans(n_clusters = aa, random_state=49).fit(df_std1)#45
labels1 = kmeans1.labels_
print(silhouette_score(df_std1, labels= labels1))
new_df = pd.DataFrame(data=df2, columns=[c[0],c[1]])
new_df['kmeans'] = labels1
df2['kmeans1'] = labels1


plt.figure(figsize=(12,8))
for i in range (aa):
    plt.scatter(new_df[c[0]][new_df.kmeans==i], new_df[c[1]][new_df.kmeans==i], c=b[i], s=100, edgecolors='green', label=str(i))
plt.legend()
plt.xlabel(c[0])
plt.ylabel(c[1])
plt.title(c[0]+' VS '+c[1])
plt.show()
```
